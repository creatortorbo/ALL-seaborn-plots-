# Styling Plot

```python
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
d = pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\csv files\\tips.csv")
d
```

```python
# here we making single box plot 
sns.boxplot(y=d["total_bill"])
plt.show() # here simple boxplot 
```

```python
# here we making the bargrapah 
sns.barplot(x ="day",y="total_bill",data = d)
plt.show() # simple bar plot as shown in below 
```

# grid parameter --->

```python
# here we created the grid  with help of seaborn (darkgrid)
sns.set_style("darkgrid")
sns.barplot(x ="day",y="total_bill",data = d)
plt.show()
```

```python
# here we created the whitegrid with the help of seaborn  
sns.set_style("whitegrid")
sns.barplot(x ="day",y="total_bill",data = d)
plt.show()
```

```python
# here we created the ticks with the help of seaborn  
sns.set_style("ticks")
sns.barplot(x ="day",y="total_bill",data = d)
plt.show() # here it just a normal graph 
```

# dedpine()---> 
it used to remove axis from the grpah

```python
sns.barplot(x ="day",y="total_bill",data = d)
sns.despine() # here we seen that removing the above axis 
plt.show()
```

# figsize()--->
it is used to increase or decrese the size of the grpah 

```python
plt.figure(figsize =(12,3))
# here hight -->3
# width ---> 12
sns.barplot(x ="day",y="total_bill",data = d) 
plt.show()
```

```python
plt.figure(figsize =(12,7))# (width ,hight)
# here hight --> 12 
# width ---> 3
sns.barplot(x ="day",y="total_bill",data = d) 
plt.show()
```

```python
plt.figure(figsize =(3,15))# (width ,hight)
sns.barplot(x ="day",y="total_bill",data = d) 
plt.show()
```

```python
# here we showing the graph in verticle 
plt.figure(figsize =(2,10))# (width ,hight)
sns.barplot(x ="day",y="total_bill",data = d) 
plt.show()
```

# set_context()--->
it is used for the fount change parameter

```python
plt.figure(figsize =(3,4))# (width ,hight)
sns.set_context("poster",font_scale=2) # here we increase the font as shown 
sns.barplot(x ="day",y="total_bill",data = d)
plt.show()
```

```python
plt.figure(figsize =(3,4))# (width ,hight)
sns.set_context("poster",font_scale=1) # here we increase the font as shown 
sns.barplot(x ="day",y="total_bill",data = d)
plt.show() # here 'poster 'it is front type 
```

```python
plt.figure(figsize =(3,4))# (width ,hight)
sns.set_context("poster",font_scale=0.2) # here we increase the font as shown 
sns.barplot(x ="day",y="total_bill",data = d)
plt.show()# we have decrease the lables  with front_scale parameter
```

```python
plt.figure(figsize =(3,4))# (width ,hight)
sns.set_context("paper",font_scale=0.2) # here we increase the font as shown 
sns.barplot(x ="day",y="total_bill",data = d)
plt.show()
```
