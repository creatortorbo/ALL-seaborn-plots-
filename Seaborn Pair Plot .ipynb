# Seaborn Pair Plot 

```python
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
# the graph is drow the number valus (total_bill,tip,size)
```

```python
d = pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\csv files\\tips.csv")
d
```

```python
sns.pairplot(d)
plt.show()
```

# hue parameter --->
it is used to make graph wst anther graph 


```python
sns.pairplot(d,hue = "sex")
plt.show()  # here the grpah is created with the respt to sex
```

```python
sns.pairplot(d,hue = "smoker")
plt.show() #here the grpah is created with the respt to smoker
```

# vars parameter ----> 
it is for we can create the grpah for any particler as shown in given below 


```python
# here we only drow  the graph of the [total_bill,tip]
sns.pairplot(d,vars = ["total_bill","tip"],hue = "smoker")
plt.show()  # here we drown the graph in which the multiople graph drwown in single graph as shown given bleow 
```

```python
d
```

```python
# here we only drow  the graph of the [total_bill,tip]
sns.pairplot(d,vars = ["tip","size"],hue = "smoker")
plt.show() # here vars is works with numaricle values 
```

# hue order parameter --->


```python

sns.pairplot(d,vars = ["tip","size"],hue = "sex",hue_order=["Female","Male"])
plt.show()
```

```python
sns.pairplot(d,vars = ["tip","size"],hue = "sex",hue_order=["Male","Female"])
plt.show()
```

# palette parameter --->
it is the color doing the graph as shown in given below 

```python
sns.pairplot(d,vars = ["tip","size"],hue = "sex",hue_order=["Male","Female"],palette= "BuGn")
plt.show()  # the BuGn ---> it the color with palette parameter 
```

```python
d
```

# x_vars parameter --->
it is for the x - axis 

```python
sns.pairplot(d,hue = "sex",hue_order=["Male","Female"],x_vars=["total_bill","tip"])
plt.show()
```

# y_vars parameter --->
it is for the y- axis the graph 

```python
sns.pairplot(d,hue = "sex",hue_order=["Male","Female"],y_vars=["total_bill","tip"])
plt.show()
```

# kind paraeter --->
it is used for changeing the graph 

```python
sns.pairplot(d,hue = "sex",hue_order=["Male","Female"],y_vars=["total_bill","tip"],kind ="reg")
plt.show()  # kind ---> ["scatter",'kde','hist','reg'] it diff type of grpah 
```

```python
# to use to change the graph as shown in given below 
sns.pairplot(d,hue = "sex",hue_order=["Male","Female"],y_vars=["total_bill","tip"],kind ="scatter")
plt.show()
```

```python
sns.pairplot(d,hue = "sex",hue_order=["Male","Female"],y_vars=["total_bill","tip"],kind ="hist")
plt.show() # here we created with king paameter make the histgram graph as shown in given below :
```

```python
sns.pairplot(d,hue = "sex",hue_order=["Male","Female"],y_vars=["total_bill","tip"],kind ="kde")
plt.show()# here use the kde graph with the kind parameter 
```

# diag_kind parameter --->
it used to change the graph in the diagonale 

```python
sns.pairplot(d,hue = "sex",hue_order=["Male","Female"],
             kind ="kde",diag_kind = "kde")
plt.show()
```

```python
sns.pairplot(d,hue = "sex",hue_order=["Male","Female"],
             kind ="kde",diag_kind = "hist")
plt.show()  # in the diagonal we craeated the histogram graph as shown in given below 
```

# markers parameter --->
it use to marks the graph with some symbole as [*,>,<] and etc.

```python
sns.pairplot(d,hue = "sex",hue_order=["Male","Female"],markers=["*","<"])
plt.show()
```
