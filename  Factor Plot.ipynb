# Factor Plot

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns 
```

```python
d = pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\csv files\\tips.csv")
d
```

```python
sns.catplot(x="size",y="tip",data = d)
plt.show() # here the factorplot as change the name as catplot 
```

# hue parameter --->

```python
sns.catplot(x="size",y="tip",data = d,hue="sex")
plt.show()
```

# kind parameter---->
it is used to change the plot 

```python
sns.catplot(x="size",y="tip",data = d,hue="sex",kind = "bar")
plt.show()  # in the kind with we have change the plot into barplot 
```

```python
# scetter plot 
sns.catplot(x="size",y="tip",data = d,hue="sex",kind = "box")
plt.show()
```

to change the color

```python
sns.catplot(x="tip",y="size",data = d,color = "g",hue = "sex",palette = "Accent",height = 2)
plt.show() # here we decrease the with help of the height parameter 
```

```python
sns.catplot(x="tip",y="size",data = d,color = "g",hue = "sex",palette = "Accent",height = 10)
plt.show()   # here we increase the graph as shown in given below 
```
