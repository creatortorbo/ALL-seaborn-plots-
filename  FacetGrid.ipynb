# FacetGrid --->Multiple plots 

```python
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
d = pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\csv files\\tips.csv")
d
```

```python
fg = sns.FacetGrid(d)
# here map to used to plot the graph 
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

# col parameter --->
here we take the clounm name and differentace the graph 
example: sex --> male and female

```python
fg = sns.FacetGrid(d,col="sex")
# here map to used to plot the graph 
fg.map(plt.scatter,"total_bill","tip")
plt.show() # here we make the graph in two parts 
```

# hue parameter --->
we can split the graph with respt to another

```python
fg = sns.FacetGrid(d,col="sex",hue="day")
# here map to used to plot the graph 
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

# add_legand()--->

```python
fg = sns.FacetGrid(d,col="sex",hue="day")
# here map to used to plot the graph 
fg.map(plt.scatter,"total_bill","tip").add_legend() 
plt.show()      ## it is used to add the legand                                     
```

```python
# here we make the multiple grpah with respt to sex
fg = sns.FacetGrid(d,col="day",hue="sex")
# here map to used to plot the graph 
fg.map(plt.scatter,"total_bill","tip").add_legend() 
plt.show() 
```

```python
# here we make the multiple grpah with respt to sex
fg = sns.FacetGrid(d,col="day",hue="smoker")
# here map to used to plot the graph 
fg.map(plt.scatter,"total_bill","tip").add_legend() 
plt.show() 
```

```python
# for bargraph 
fg = sns.FacetGrid(d,col="day",hue="smoker") 
fg.map(plt.bar,"total_bill","tip").add_legend() 
plt.show()
```

```python
# for bargraph 
fg = sns.FacetGrid(d,col="smoker",hue="day") 
fg.map(plt.bar,"total_bill","tip").add_legend() 
plt.show()
```

# palette parameter--->

```python
# for bargraph 
fg = sns.FacetGrid(d,col="smoker",hue="day",palette ="Accent" ) 
fg.map(plt.bar,"total_bill","tip").add_legend() 
plt.show()
```

# edgecolor parameter --->
it used color the edge of the ploted graph 

```python
# for bargraph 
fg = sns.FacetGrid(d,col="smoker",hue="day",palette ="Accent" ) 
fg.map(plt.bar,"total_bill","tip",edgecolor="g").add_legend() 
plt.show()   # here we color the edge in the green color
```

```python
# for bargraph 
fg = sns.FacetGrid(d,col="smoker",hue="day",palette ="Accent" ) 
fg.map(plt.bar,"total_bill","tip",edgecolor="r").add_legend() 
plt.show() 
```
