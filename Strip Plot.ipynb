# Strip Plot

```python
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
d = pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\csv files\\tips.csv")
d
```

```python
sns.stripplot(x = "day",y="total_bill",data=d )
plt.show()
```

```python

sns.stripplot(x = "day",y="total_bill",data=d, color = "b")
plt.show()
```

# hue parameter --->

```python
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex")
plt.show()
```

# palettle parameter --->

```python
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex",palette="rocket_r")
plt.show() #  with the help of this parameter in which is we change the color of the graph 
```

```python
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex",palette="summer_r")
plt.show() # here we change the color with the palette parameter 
```

# linewidth parameter --->
it is used for the line width increase or decrease 

```python
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex",palette="rocket_r",
              linewidth=2)
plt.show() 
```

```python
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex",palette="rocket_r",
              linewidth=3)
plt.show() # here we increase the width in the graph 
```

# edgecolor parameter --->

```python
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex",palette="rocket_r",
              linewidth=3,edgecolor = "y")
plt.show() # here we given the the edge color is yellow as shown in graph 
```

```python
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex",palette="rocket_r",
              linewidth=1,edgecolor = "b")
plt.show()
```

# jitter parameter --->
it is distibuted the like like the scetter plot 

```python
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex",palette="rocket_r",
              linewidth=1,edgecolor = "b",jitter=1)
plt.show()
```

```python
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex",palette="rocket_r",
              linewidth=1,edgecolor = "b",jitter=5)
plt.show() # here due to this it can distubeded in the graph 
```

# size parameter --->
it is for the increase and decrease the size of bars 

```python
# here the circle is given we can change the size of the circle 
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex",palette="rocket_r",
              linewidth=1,edgecolor = "b",size = 10) # here size is 10
plt.show()
```

```python
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex",palette="rocket_r",
              linewidth=1,edgecolor = "b",size = 1) # here size is 1
plt.show() # due to the size parameter given is 1 and the 
```

# marker parameter --->
it is use to set the marker 

```python
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex",palette="rocket_r",
              linewidth=1,edgecolor = "b",size = 5,marker = "<") # here size is 1
plt.show()
```

# alpha parameter --->
it is used to decrease and increase the light 

```python
sns.stripplot(x = "day",y="total_bill",data=d, color = "y",hue = "sex",palette="rocket_r",
              linewidth=1,edgecolor = "b",size = 5,marker = "<",alpha=0.4) # here size is 1
plt.show() #  here we decrease the color of the bars
```

# single stripplot in x -axis 

```python
sns.stripplot(x=d["total_bill"],color = "g")
plt.show()
```
