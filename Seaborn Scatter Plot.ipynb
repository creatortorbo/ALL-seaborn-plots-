# style parameter ---> 

```python
# here we using the style parameter for the changing the styles in the graph 
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = d,hue = "sex",style = "sex")
plt.show() 

```

# size parameter --->
it used to change the size of the plots 

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = d,hue = "sex",style = "sex",size="sex" )
plt.show() 

```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = d,hue = "sex",style = "sex",sizes=(10,40))
plt.show() 
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = d,hue = "sex",style = "sex",size="sex",sizes= (100,40) )
plt.show()             # here we increase the size of the Female                                # it is the FEMALE,MALE 
```

# palette parameter --->
it is for the colors change in this graph

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = d,hue = "sex",style = "sex",
                size="sex",sizes= (100,40),palette = "gist_ncar" )
plt.show()   
```

# alpha parameter --->
it is used to decrease or icrease the light of the color

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = d,hue = "sex",style = "sex",
                size="sex",sizes= (100,40),palette = "gist_ncar",alpha = 0.5 )
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = d,hue = "sex",style = "sex",
                size="sex",sizes= (100,40),palette = "gist_ncar",alpha = 0.1 )
plt.show() # here we decrease the color of the light                      alpah ---> 0 b/w 1
```

# markers parameter --->
it is used to make a marks

```python
m = {"MALE":"*","FEMALE":"^"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data = d,hue = "sex",style = "sex",
                size="sex",sizes= (80,70),palette = "gist_ncar",markers = m )
plt.show() 
```
