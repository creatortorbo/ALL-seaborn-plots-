# Box Plot 

```python
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns 
```

```python
d = pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\csv files\\tips.csv")
d
```

```python
sns.set(style="whitegrid") # here we created the whitegrid in given graph 
```

```python
sns.boxplot(x = "day",y="total_bill",data=d)
plt.show()
```

# hue parameter --->

```python
sns.boxplot(x = "day",y="total_bill",data=d,hue="sex")
plt.show()
```

```python
# hue parameter --->
sns.boxplot(x = "day",y="total_bill",data=d,hue="time")
plt.show()
```

# order parameter --->
it is used to set order 

```python
order_1=["Fri","sun","Thur","sat"] # here we set the order in given below 
sns.boxplot(x = "day",y="total_bill",data=d,hue="time",order = order_1)
plt.show()
```

# showmeans parameter ---> 
it is used for the 

```python
order_1=["Fri","sun","Thur","sat"]
sns.boxplot(x = "day",y="total_bill",data=d,order = order_1,showmeans = True)
plt.show() # here the arroy is our mean as shown in given bleow 
```

# meanprops parameter --->
it is using the maean proptices 

```python
order_1=["Fri","sun","Thur","sat"]
sns.boxplot(x = "day",y="total_bill",data=d,order = order_1,
            showmeans = True,meanprops = {"marker":"+","markeredgecolor":"y"})
plt.show()
```
