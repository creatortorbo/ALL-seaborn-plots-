# Seaborn Count Plot

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt 
```

```python
d = pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\csv files\\tips.csv")
d
```

```python
sns.countplot(x ="sex",data= d)
plt.show()
```

```python
order_1 =["Male","Female"] 
sns.countplot(x ="sex",data= d,order=order_1)
plt.show()
```

```python
# In the counter plot we counting the number 
order_1 =["Yes","No"] 
sns.countplot(x ="smoker",data= d,order=order_1)
plt.show()
```

```python
# comparing the grpah with bargraph 
sns.barplot(x="sex",y ="size",data =d)  # barplot is this in graph 
plt.show()
```

```python
order_1 =["Male","Female"] 
sns.countplot(x ="sex",data= d,order=order_1,hue = "smoker")
plt.show()
```

# horignatal graph 

```python
order_1 =["Male","Female"] 
sns.countplot(y ="sex",data= d,order=order_1,hue = "smoker")
plt.show() # here we are make grpah in the horigntial form as shown in given below 
```

# palette parameter --->
it is for the color parameter 

```python
order_1 =["Male","Female"] 
sns.countplot(y ="sex",data= d,order=order_1,hue = "smoker",palette = "bwr")
plt.show()
```

```python
# normal color parameter is used in the graph is also 
order_1 =["Male","Female"] 
sns.countplot(y ="sex",data= d,order=order_1,hue = "smoker",color = "y")
plt.show()  # one is simple is wight and onother is yellow 
```

# saturation parameter --->
it of color decrease parameter(brightness) 

```python
# normal color parameter is used in the graph is also 
order_1 =["Male","Female"] 
sns.countplot(x ="sex",data= d,order=order_1,hue = "smoker",color = "r",saturation = 0.1)
plt.show()
```
