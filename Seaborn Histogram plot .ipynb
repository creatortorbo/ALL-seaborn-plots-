# Seaborn Histogram Plot Method in Python

```python
import pandas as pd
import seaborn as sns 
from matplotlib import pyplot as plt 
```

```python
d = pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\csv files\\penguins.csv")
d
```

```python
sns .displot(d["bill_depth_mm"])
plt.show()
```

```python
sns .displot(d["flipper_length_mm"],color = "r")
plt.show()
```

# bins parameter ----> it used to arrage the number to make the range 

```python
sns .displot(d["flipper_length_mm"],color = "r",bins=[170,180,190,200,210,220,230])
plt.show()
# here we used bins parameter  --> due to this we can data are orgines 
```

```python
sns .displot(d["flipper_length_mm"],color = "r")
plt.show()
```

# kde parameter ----> 
Kernel Density Estimation it used for drowing the line on the graph 

```python
sns .displot(d["flipper_length_mm"],color = "r",kde=True)
plt.show()
```

```python
sns .displot(d["flipper_length_mm"],color = "g",kde=True)
plt.show()
```

# rug parameter 
To showing the kde in axis 

```python
sns .displot(d["flipper_length_mm"],color = "g",kde=True,rug =True )
plt.show()
```

# log_seale parameter --->
it used to convert the table into log table as shown given bellow 

```python
sns .displot(d["flipper_length_mm"],color = "g",kde=True,rug =True,log_scale= True  )
plt.show()
# we can converted into the log data
```
