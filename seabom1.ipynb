# Seaborn Line Plot Method

```python
# here we using the large amount of data with of seaborn 
import seaborn as sns
from matplotlib import pyplot as plt 
import pandas as pd
```

```python
var = [1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]

x_1 = pd.DataFrame({"var":var,"var_1":var_1})

sns.lineplot(x="var",y="var_1",data=x_1)


plt.show()
```

```python
# in matplotlib use of line plot like this 
plt.plot(var,var_1)
plt.show()
```
