```python
sns.heatmap(var) # here used the seaborn 
plt.show()
# it moves dark to light 
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1  #  new 2 d array 
```

```python
# now we making the heatmap 
sns.heatmap(var1,vmin= 0,vmax=20,cmap="gist_heat",annot = True)
plt.show()
```

```python
# creating the 2 d array
ar = np.array([["a0","a1","a2","a3","a4"],
              ["b0","b1","b2","b3","b4"]])
ar  # the given below the 2D array 
```

# fmt parameter ---->
it is used of the convet diff datatype 

```python
sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s") # fmt --> string datatype
plt.show()
```

# fontsize and color parameter ----->
it is for change the color and fontsize 

```python
# data we have with dict 
y={"fontsize":10,"color":"r"}
sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s",annot_kws= y) # fmt --> string datatype
plt.show()
```

```python
# data we have with dict 
y={"fontsize":20,"color":"y"} # here we have cahnge the color and fontsize 
sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s",annot_kws= y) # fmt --> string datatype
plt.show()
```

# linewidth parameter --->
 in the box space line

```python
# data we have with dict 
y={"fontsize":20,"color":"y"} 
sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s",annot_kws= y,linewidth = 4) # line gap in the graph due to this parameter 
plt.show()  # linespace 
```

```python
# data we have with dict 
y={"fontsize":20,"color":"y"} 
sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s",annot_kws= y,linewidth = 10)  
plt.show()
```

```python
# data we have with dict 
y={"fontsize":20,"color":"y"} 
sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s",annot_kws= y,linewidth = 20)  
plt.show()  # here we created big space in the graph with helps of the linewidth
```

# linecolor parameter --->
it for the linecolor in the grpah

```python
y={"fontsize":20,"color":"y"} 
sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s",annot_kws= y,linewidth = 20 ,linecolor= "y")  
plt.show()
```

```python
y={"fontsize":20,"color":"y"} 
sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s",annot_kws= y,linewidth = 20 ,linecolor= "b")  
plt.show()
```

```python
y={"fontsize":20,"color":"y"} 
sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s",annot_kws= y,linewidth = 20 ,linecolor= "g")  
plt.show()
```

# cbar ---> to removing the color bar 

```python
y={"fontsize":20,"color":"y"} 
sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s",annot_kws= y,linewidth = 20 ,linecolor= "g",cbar =False)  
plt.show()  # here remove the color bar form the grpah 
```

# xticklabels parameter ---->
it for the removing the lables form the x - axis 

```python
# To removing the lables and  x - axis 
y={"fontsize":20,"color":"y"} 
sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s",annot_kws= y,linewidth = 20 ,linecolor= "g",cbar =False,xticklabels = False)  
plt.show()
```

```python
# To removing the lables and  y - axis 
y={"fontsize":20,"color":"y"} 
sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s",annot_kws= y,linewidth = 20 ,linecolor= "g",cbar =False,xticklabels = False,yticklabels = False)  
plt.show()
```

```python
# we whant to be creating the lables in the grpah 
# To removing the lables and  y - axis 
y={"fontsize":20,"color":"y"} 
sb  = sns.heatmap(var1,vmin= 0,vmax=10,cmap="gist_heat",
            annot = ar,fmt= "s",annot_kws= y,linewidth = 20 ,linecolor= "g",cbar =False,
            xticklabels = False,yticklabels = False)
sb.set(xlabel = "python",ylabel="numpy")

plt.show()
```
