# Violin Plot Kya Hai 

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
# it the maltiple graph in the same graph as shown given below 
```

```python
d = pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\csv files\\tips.csv")
d
```

```python
 # it is the sample basic violineplot 
sns.violinplot(x="day",y="total_bill",data = d)
plt.show()
```

# hue parameter --->

```python
sns.violinplot(x="day",y="total_bill",data = d,hue = "time")
plt.show()
```

# linewidth parameter ---> 
it the inceasing the line width 

```python
sns.violinplot(x="day",y="total_bill",data = d,hue = "time",linewidth=4)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data = d,hue = "time",linewidth=1)
plt.show()
```

# palette parameter --->
it is used for the color change parameter 

```python
sns.violinplot(x="day",y="total_bill",data = d,hue = "time",linewidth=1,palette ="Dark2")
plt.show()
```

# order parameter --->
we can convert diff order

```python
 # it is the sample basic violineplot 
order_1=["Lunch","Dinner"]
sns.violinplot(x="time",y="total_bill",data = d,order=order_1)
plt.show()
```

# saturation parameter --->
it is the used for colors light(brigthness)

```python
order_1=["Lunch","Dinner"]
sns.violinplot(x="time",y="total_bill",data = d,order=order_1,saturation=0.5)
plt.show()
```

```python
order_1=["Lunch","Dinner"]
sns.violinplot(x="time",y="total_bill",data = d,order=order_1,saturation=0.1)
plt.show()
```

```python
order_1=["Lunch","Dinner"]
sns.violinplot(x="time",y="total_bill",data = d,order=order_1,saturation=0.0)
plt.show() # here we remove the color by the saturation parameter as shown given below 
```

# color parameter --->
it is used to color graph as shown in given below 

```python
order_1=["Lunch","Dinner"]
sns.violinplot(x="time",y="total_bill",data = d,order=order_1,color = "r" )
plt.show() 
```

# split parameter --->
it is to split the graph 

```python

sns.violinplot(x="day",y="total_bill",data = d,hue = "sex",split= True )
plt.show()  # here we are diff b/w male and female
```

# scale prameter --->
we change the width 

```python
sns.violinplot(x="day",y="total_bill",data = d,hue = "sex",split= True,scale = "width" )
plt.show()
```

# horigntal graph --->
with help of this we drow in horiginatl graph with this 

```python
sns.violinplot(x="total_bill",y="day",data = d,hue = "sex",split= True,scale = "width" )
plt.show()
```

# inner parameter --->
it is used change the plot 

```python
order_1=["Lunch","Dinner"]
sns.violinplot(x="time",y="total_bill",data = d,order=order_1,inner= "box")
plt.show()   # in the inner parameter we use the box plot here 
```

```python
order_1=["Lunch","Dinner"]
sns.violinplot(x="time",y="total_bill",data = d,order=order_1,inner= "point")
plt.show()
```

```python
order_1=["Lunch","Dinner"]
sns.violinplot(x="time",y="total_bill",data = d,order=order_1,inner= "quart")
plt.show()                                   # {"quartile","point","stick",None}
```
