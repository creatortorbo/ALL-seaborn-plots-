# Seaborn BarPlot

```python
import pandas as pd
import seaborn as sns 
from matplotlib import pyplot as plt 
```

```python
d = pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\csv files\\penguins.csv")
d
```

```python
sns.barplot(x =d.island,y=d.bill_length_mm)
plt.title("python",fontsize= 20) # here we use font size for size
plt.show()
```

```python
sns.barplot(x ="island",y="bill_length_mm",data=d)
plt.title("python",fontsize= 20) # here we use font size for size
plt.show()
```

# hue parameter ---> 
it drow the graph with resepts to another 

```python
sns.barplot(x ="island",y="bill_length_mm",data=d,hue = "sex")
plt.title("python",fontsize= 20) # here we drown the with respects to sex as shown given below :
plt.show()
```

# order parameter ---->
it can change the order of the given data 

```python
order_1=["Biscoe","Dream","Torgersen"]
sns.barplot(x ="island",y="bill_length_mm",data=d,hue = "sex",order = order_1)
plt.title("python",fontsize= 20) 
plt.show()
```

```python
order_1=["Dream","Biscoe","Torgersen"]# here we are changeing the order of the given data 
sns.barplot(x ="island",y="bill_length_mm",data=d,hue = "sex",order = order_1)
plt.title("python",fontsize= 20) 
plt.show()
```

# hue _order --->
In the legend we change the order and with the helps of the hue parameter 

```python
order_1=["Dream","Biscoe","Torgersen"]# here we are changeing the order of the given data 
sns.barplot(x ="island",y="bill_length_mm",data=d,hue = "sex",
            order = order_1,hue_order = ["FEMALE","Male"])
plt.title("python",fontsize= 20) 
plt.show()
```

# ci parameter ---> 
it is for the incresing and decresing line jn the graph  

```python
order_1=["Dream","Biscoe","Torgersen"]# here we are changeing the order of the given data 
sns.barplot(x ="island",y="bill_length_mm",data=d,hue = "sex",
            order = order_1,ci=10)
plt.title("python",fontsize= 20) 
plt.show()
```

```python
order_1=["Dream","Biscoe","Torgersen"]# here we are changeing the order of the given data 
sns.barplot(x ="island",y="bill_length_mm",data=d,hue = "sex",
            order = order_1,ci=100,n_boot = 2)
plt.title("python",fontsize= 20) 
plt.show()
```

# dodge parameter---->
it is used for one bar laying on another bar 

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x ="island",y="bill_length_mm",data=d,hue = "sex",
            order = order_1,palette="icefire",saturation = 10,errcolor = "r", dodge = False) 
plt.title("python",fontsize= 20, )    # here we are laying on one another 
plt.show()
```

# alpah parameter --->
it is used for the light (0-1)

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x ="island",y="bill_length_mm",data=d,hue = "sex",
            order = order_1,palette="icefire",saturation = 100,errcolor = "r", dodge = False,alpha = 0.7) # decrease the light 
plt.title("python",fontsize= 20, )    
plt.show()
```
