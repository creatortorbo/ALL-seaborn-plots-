# Seaborn Line Plot Method in Python

```python
from matplotlib import pyplot as plt 
import seaborn as sns 
import pandas as pd
```

# our matplotlib graph 

```python
data = [1,2,3,4,5,6,7]
data_1 = [2,4,5,6,4,3,7]
plt.plot(data,data_1)
plt.show()
```

# seaborn graph 

# taking data frame csv file 


```python
data = [1,2,3,4,5,6,7]
data_1 = [2,4,5,6,4,3,7]

x_1 = pd.DataFrame({"data":data,"data_1":data_1}) # here we sue  of dataframe 
sns.lineplot(x = "data",y="data_1",data = x_1)
plt.show()
```

```python
# deirect use without dataframe in this 
data = [1,2,3,4,5,6,7]
data_1 = [2,4,5,6,4,3,7]
x_1 = sns.lineplot(x = data,y = data_1)
plt.show()
```

# hue parameter --->powerfull data ka respest graph 

```python
# here using the power full
d_1  = sns.lineplot(x = "bill_length_mm",y="bill_depth_mm",data = d,hue = "sex")
plt.show()
# we make the with r to sex (in power full data )
```

# size parameter ---> we can change the size with it 


```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",size =5)
plt.show()
# size of ploted line we can
```

```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",size =100)
plt.show()
# size of ploted line we can increase and decrease the breath
```

# style parameter ---> it used for the doted line 

```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",style = "sex")
plt.show()
```

# palette parameter ---> it use to color the graph line 

```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",style = "sex",palette = "Accent")
plt.show()
```

# Marker parameter ---> it used to mark the line above 

```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",style = "sex",palette = "Accent",markers= ["o",">"])
plt.show()# o is at the fist line graph and > it the second graph 
```

```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",style = "sex",palette = "Accent",markers= ["^","*"])
plt.show()# o is at the fist line graph and > it the second graph 
```

# dashes  parameter ----> it used to remove the doted line from the graph 

```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",style = "sex",palette = "Accent",markers= ["o",">"],dashes = True)
plt.show()# o is at the fist line graph and > it the second graph 
```

```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",style = "sex",palette = "Accent",markers= ["o",">"],dashes= False)
plt.show()# o is at the fist line graph and > it the second graph 
# in dashes parameter ----> when false with we can remove the doted line 
```

# In seaborn we remove the legend with legend parameter 

```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",style = "sex",palette = "Accent",
             markers= ["o",">"],dashes= False,legend = False)
plt.show()# o is at the fist line graph and > it the second graph 
```

# In the legend parameter we have diff 

```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",style = "sex",palette = "Accent",
             markers= ["o",">"],dashes= False,legend = "f")
plt.show()
```

```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",style = "sex",palette = "Accent",
             markers= ["o",">"],dashes= False,legend = "full")
plt.show()
```

```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",style = "sex",palette = "Accent",
             markers= ["o",">"],dashes= False,legend = 'auto')
plt.show()
```

# grid() 
it used with matplotlib 

```python
sns.lineplot(x ="bill_length_mm",y = "bill_depth_mm",data = d, hue = "sex",style = "sex",palette = "Accent",
             markers= ["o",">"],dashes= False,legend = 'auto')
plt.grid()
plt.title("python")
plt.show()
```
